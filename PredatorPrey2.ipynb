```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter model Lotka-Volterra
alpha = 1.2   # Laju pertumbuhan mangsa
beta = 0.1   # Laju kematian mangsa karena predator
gamma = 1   # Laju kematian predator
delta = 0.075  # Laju pertumbuhan predator karena mangsa

# Kondisi awal
x0 = 40  # Populasi mangsa awal
y0 = 9   # Populasi predator awal

# Waktu simulasi
T = 11   # Total waktu
dt = 0.1  # Interval waktu
n = int(T / dt)  # Jumlah langkah waktu

# Inisialisasi array untuk hasil simulasi
x = np.zeros(n)
y = np.zeros(n)
# Initialize dx and dy as arrays
dx = np.zeros(n)
dy = np.zeros(n)
t = np.linspace(0, T, n)

# Kondisi awal
x[0] = x0
y[0] = y0

# Simulasi menggunakan metode Euler
for i in range(1, n):
    dx[i] = alpha * x[i-1] - beta * x[i-1] * y[i-1]
    dy[i] = delta * x[i-1] * y[i-1] - gamma * y[i-1]
    x[i] = x[i-1] + dx[i] * dt # Use dx[i] and dy[i] to access the current calculated values
    y[i] = y[i-1] + dy[i] * dt # Use dx[i] and dy[i] to access the current calculated values

print(x)
print(y)
print(dx)
print(dy)

# Plot hasil simulasi: x(t) dan y(t) terhadap t
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x, 'b', label='Mangsa (x)')
plt.plot(t, y, 'r', label='Predator (y)')
plt.xlabel('Waktu')
plt.ylabel('Populasi')
plt.title('Populasi Mangsa dan Predator terhadap Waktu')
plt.legend()
plt.grid(False)

# Plot hasil simulasi: y terhadap x (fase plot)
plt.subplot(2, 1, 2)
plt.plot(x, y, 'g')
plt.xlabel('Populasi Mangsa (x)')
plt.ylabel('Populasi Predator (y)')
plt.title('Plot Populasi Mangsa vs Predator')
plt.grid(False)

plt.tight_layout()
plt.show()
```
