```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter untuk model Lotka-Volterra
alpha = 1.2   # Tingkat pertumbuhan mangsa
beta = 0.1   # Tingkat interaksi mangsa-predator
delta = 0.075  # Efisiensi predator memakan mangsa
gamma =  1 # Tingkat kematian predator

# Waktu simulasi
T = 11     # Jumlah iterasi (waktu)
dt = 0.1     # Langkah waktu
time = np.arange(0, T*dt, dt)  # Vektor waktu

# Inisialisasi populasi awal
x = np.zeros(len(time))  # Populasi mangsa
y = np.zeros(len(time))  # Populasi predator
x[0] = 40  # Populasi awal mangsa
y[0] = 9   # Populasi awal predator

# Metode Euler untuk menyelesaikan persamaan diferensial Lotka-Volterra
for t in range(1, len(time)):
    dx_dt = alpha * x[t-1] - beta * x[t-1] * y[t-1]
    dy_dt = delta * x[t-1] * y[t-1] - gamma * y[t-1]

    x[t] = x[t-1] + dx_dt * dt
    y[t] = y[t-1] + dy_dt * dt

print(x)
print(y)
print(dx_dt)
print(dy_dt)

# Plot hasil simulasi
plt.figure(figsize=(10,6))
plt.plot(time, x, label='Prey (Mangsa)')
plt.plot(time, y, label='Predator')
plt.title('Simulasi Model Lotka-Volterra (Predator-Prey)')
plt.xlabel('Waktu')
plt.ylabel('Populasi')
plt.legend()
plt.grid(False)
plt.show()
```
